```python
# =======================================================
# 2ë‹¨ê³„: ë°ì´í„° íƒìƒ‰(EDA) ë° ë‹¤ì¤‘ê³µì„ ì„± ì§„ë‹¨
# =======================================================

print("ğŸ”„ 2ë‹¨ê³„ ë¶„ì„ ì‹œì‘: ë°ì´í„° íƒìƒ‰ ë° ë‹¤ì¤‘ê³µì„ ì„± ì§„ë‹¨")

# --- 1. ëª©í‘œ ë³€ìˆ˜(y) ë¶„í¬ í™•ì¸ ---
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.histplot(y, kde=True, bins=30)
plt.title('ëª©í‘œ ë³€ìˆ˜ "ìƒê¶ŒíšŒì „ìœ¨" ë¶„í¬', fontsize=15)
plt.xlabel('ìƒê¶ŒíšŒì „ìœ¨')
plt.ylabel('ë¹ˆë„')

plt.subplot(1, 2, 2)
sns.boxplot(x=y)
plt.title('ëª©í‘œ ë³€ìˆ˜ "ìƒê¶ŒíšŒì „ìœ¨" ë°•ìŠ¤í”Œë¡¯', fontsize=15)
plt.xlabel('ìƒê¶ŒíšŒì „ìœ¨')
plt.grid(True, axis='x')

plt.suptitle('1. ëª©í‘œ ë³€ìˆ˜(y) íƒìƒ‰', fontsize=18, fontweight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# --- 2. ì£¼ìš” íŠ¹ì„±(X)ê³¼ ëª©í‘œ ë³€ìˆ˜(y) ê°„ì˜ ìƒê´€ê´€ê³„ ë¶„ì„ ---
print("\nğŸ“Š 'ìƒê¶ŒíšŒì „ìœ¨'ê³¼ ìƒê´€ê´€ê³„ê°€ ë†’ì€ ìƒìœ„ íŠ¹ì„± ë¶„ì„")

# ìƒê´€ê³„ìˆ˜ ê³„ì‚°
correlations = X.corrwith(y).sort_values(ascending=False)
corr_df = pd.DataFrame({'correlation': correlations})
corr_df['abs_correlation'] = corr_df['correlation'].abs()
corr_df_sorted = corr_df.sort_values('abs_correlation', ascending=False)

print("\n--- ìƒìœ„ 10ê°œ ìƒê´€ê´€ê³„ íŠ¹ì„± ---")
display(corr_df_sorted.head(10))

# ìƒìœ„ 10ê°œ íŠ¹ì„± ì‹œê°í™”
plt.figure(figsize=(10, 8))
top_10_features = corr_df_sorted.head(10)
sns.barplot(x=top_10_features['correlation'], y=top_10_features.index)
plt.title('2. "ìƒê¶ŒíšŒì „ìœ¨"ê³¼ ìƒê´€ê´€ê³„ê°€ ë†’ì€ Top 10 íŠ¹ì„±', fontsize=15, fontweight='bold')
plt.xlabel('ìƒê´€ê³„ìˆ˜')
plt.ylabel('íŠ¹ì„±ëª…')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()

# --- 3. ë‹¤ì¤‘ê³µì„ ì„± ì§„ë‹¨ ---
print("\nğŸ”¥ ì¤‘ìš”: ë‹¤ì¤‘ê³µì„ ì„±(Multicollinearity) ì§„ë‹¨")
print("   - ìƒê´€ê´€ê³„ê°€ ë†’ì€ ìƒìœ„ íŠ¹ì„±ë“¤ì´ ì„œë¡œ ì–¼ë§ˆë‚˜ ê°•í•˜ê²Œ ì—°ê´€ë˜ì–´ ìˆëŠ”ì§€ í™•ì¸í•©ë‹ˆë‹¤.")

# ìƒìœ„ 10ê°œ íŠ¹ì„± ê°„ì˜ ìƒê´€ê´€ê³„ ë§¤íŠ¸ë¦­ìŠ¤ ê³„ì‚°
top_10_feature_names = top_10_features.index.tolist()
top_features_corr_matrix = X[top_10_feature_names].corr()

# íˆíŠ¸ë§µìœ¼ë¡œ ì‹œê°í™”
plt.figure(figsize=(10, 8))
sns.heatmap(top_features_corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('3. Top 10 íŠ¹ì„± ê°„ì˜ ìƒê´€ê´€ê³„ íˆíŠ¸ë§µ', fontsize=15, fontweight='bold')
plt.show()

print("\n--- ì§„ë‹¨ ê²°ê³¼ ---")
print("íˆíŠ¸ë§µì—ì„œ ë¶‰ì€ìƒ‰ì´ ì§„í• ìˆ˜ë¡(1ì— ê°€ê¹Œìš¸ìˆ˜ë¡) ë³€ìˆ˜ ê°„ ìƒê´€ê´€ê³„ê°€ ë§¤ìš° ë†’ë‹¤ëŠ” ì˜ë¯¸ì…ë‹ˆë‹¤.")
print("ì˜ˆë¥¼ ë“¤ì–´, 'ì›”í‰ê· ë§¤ì¶œê¸ˆì•¡'ê³¼ 'í‰ê· ê±°ë˜ê¸ˆì•¡' ê°™ì€ ë³€ìˆ˜ë“¤ì€ ê±°ì˜ ë™ì¼í•œ ì •ë³´ë¥¼ ë‹´ê³  ìˆì–´ ë‹¤ì¤‘ê³µì„ ì„± ë¬¸ì œê°€ ë°œìƒí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.")
print("ì´ëŠ” ëª¨ë¸ì˜ ì•ˆì •ì„±ì„ í•´ì¹˜ê³  íŠ¹ì • ë³€ìˆ˜ì˜ ì¤‘ìš”ë„ë¥¼ ì™œê³¡ì‹œí‚¤ëŠ” ì£¼ìš” ì›ì¸ì´ ë©ë‹ˆë‹¤.")
print("\nâœ… 2ë‹¨ê³„ ë¶„ì„ ì™„ë£Œ. ë‹¤ìŒ ë‹¨ê³„ì—ì„œëŠ” ì´ ë¬¸ì œë¥¼ í•´ê²°í•˜ê¸° ìœ„í•œ íŠ¹ì„± ì„ íƒì„ ì§„í–‰í•©ë‹ˆë‹¤.")
```

```python
# =======================================================
# 3ë‹¨ê³„: ì§€ëŠ¥ì ì¸ íŠ¹ì„± ì„ íƒ ë° ëœë¤í¬ë ˆìŠ¤íŠ¸ ëª¨ë¸ë§
# =======================================================
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

print("ğŸ”„ 3ë‹¨ê³„ ë¶„ì„ ì‹œì‘: ì§€ëŠ¥ì ì¸ íŠ¹ì„± ì„ íƒ ë° ëœë¤í¬ë ˆìŠ¤íŠ¸ ëª¨ë¸ë§")

# --- 1. ë‹¤ì¤‘ê³µì„ ì„± ì œì–´ë¥¼ ìœ„í•œ íŠ¹ì„± ì„ íƒ ---
print("\nğŸ”§ ë‹¤ì¤‘ê³µì„ ì„± ì œì–´ë¥¼ ìœ„í•œ íŠ¹ì„± ì„ íƒ ì¤‘...")

# ì´ì „ì— ê³„ì‚°ëœ ìƒê´€ê´€ê³„ ìˆœì„œ(corr_df_sorted)ë¥¼ ê·¸ëŒ€ë¡œ í™œìš©
features_to_consider = corr_df_sorted.index.tolist()
selected_features = []
correlation_threshold = 0.8 # ìƒê´€ê³„ìˆ˜ 0.8 ì´ìƒì´ë©´ ë†’ì€ ê²ƒìœ¼ë¡œ ê°„ì£¼

for feature in features_to_consider:
    if not selected_features:
        selected_features.append(feature)
        continue

    # ì„ íƒëœ íŠ¹ì„±ë“¤ê³¼ì˜ ìƒê´€ê´€ê³„ë¥¼ ê³„ì‚°
    correlations_with_selected = X[selected_features].corrwith(X[feature])

    # ìƒê´€ê´€ê³„ì˜ ì ˆëŒ“ê°’ ì¤‘ ìµœëŒ€ê°’ì´ ì„ê³„ê°’ë³´ë‹¤ ë‚®ìœ¼ë©´ í•´ë‹¹ íŠ¹ì„± ì¶”ê°€
    if correlations_with_selected.abs().max() < correlation_threshold:
        selected_features.append(feature)

# ë„ˆë¬´ ë§ì€ íŠ¹ì„±ì„ ì‚¬ìš©í•˜ì§€ ì•Šê¸° ìœ„í•´ ìƒìœ„ 15ê°œë§Œ ìµœì¢… ì„ íƒ
final_features = selected_features[:15]

print(f"\nâœ… ë‹¤ì¤‘ê³µì„ ì„±ì„ ì œì–´í•˜ì—¬ ìµœì¢… ì„ íƒëœ íŠ¹ì„± ({len(final_features)}ê°œ):")
print(final_features)


# --- 2. ëœë¤í¬ë ˆìŠ¤íŠ¸ ëª¨ë¸ í•™ìŠµ ---
print("\nğŸŒ² ëœë¤í¬ë ˆìŠ¤íŠ¸ ëª¨ë¸ í•™ìŠµ ë° íŠ¹ì„± ì¤‘ìš”ë„ ë¶„ì„...")

# ì„ íƒëœ íŠ¹ì„±ìœ¼ë¡œ ë°ì´í„° ì¬êµ¬ì„±
X_selected = X[final_features]

# ë°ì´í„° ë¶„í•  (í›ˆë ¨ 80%, ê²€ì¦ 20%)
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# ëœë¤í¬ë ˆìŠ¤íŠ¸ ëª¨ë¸ ìƒì„± ë° í•™ìŠµ (ê³¼ì í•© ë°©ì§€ë¥¼ ìœ„í•´ í•˜ì´í¼íŒŒë¼ë¯¸í„° ì¡°ì ˆ)
rf_model = RandomForestRegressor(
    n_estimators=100,      # íŠ¸ë¦¬ ê°œìˆ˜
    max_depth=8,           # ìµœëŒ€ ê¹Šì´ ì œí•œ
    min_samples_split=10,  # ë…¸ë“œ ë¶„í• ì„ ìœ„í•œ ìµœì†Œ ìƒ˜í”Œ ìˆ˜
    min_samples_leaf=5,    # ë¦¬í”„ ë…¸ë“œì˜ ìµœì†Œ ìƒ˜í”Œ ìˆ˜
    random_state=42,
    n_jobs=-1              # ëª¨ë“  CPU ì½”ì–´ ì‚¬ìš©
)
rf_model.fit(X_train, y_train)

# --- 3. ëª¨ë¸ ì„±ëŠ¥ í‰ê°€ ë° íŠ¹ì„± ì¤‘ìš”ë„ í™•ì¸ ---
# ê²€ì¦ ë°ì´í„°ë¡œ ì˜ˆì¸¡ ë° ì„±ëŠ¥ í‰ê°€
y_pred = rf_model.predict(X_test)
r2 = r2_score(y_test, y_pred)
print(f"\nğŸ“Š ëª¨ë¸ ì„±ëŠ¥ í‰ê°€ (ê²€ì¦ ë°ì´í„°): RÂ² = {r2:.4f}")
if r2 > 0.5:
    print("   - ëª¨ë¸ì´ 'ìƒê¶ŒíšŒì „ìœ¨'ì˜ ë³€ë™ì„±ì„ ìƒë‹¹íˆ ì˜ ì„¤ëª…í•˜ê³  ìˆìŠµë‹ˆë‹¤.")
else:
    print("   - ëª¨ë¸ ì„±ëŠ¥ì´ ë‹¤ì†Œ ë‚®ìŠµë‹ˆë‹¤. ì¶”ê°€ì ì¸ íŠ¹ì„± ê³µí•™ì´ë‚˜ ëª¨ë¸ íŠœë‹ì„ ê³ ë ¤í•´ë³¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.")


# íŠ¹ì„± ì¤‘ìš”ë„ ì¶”ì¶œ
feature_importances = pd.DataFrame({
    'feature': final_features,
    'importance': rf_model.feature_importances_
}).sort_values('importance', ascending=False)

print("\n--- ğŸ† ëœë¤í¬ë ˆìŠ¤íŠ¸ê°€ ì„ ì •í•œ Top 10 ì¤‘ìš” íŠ¹ì„± ---")
display(feature_importances.head(10))

# íŠ¹ì„± ì¤‘ìš”ë„ ì‹œê°í™”
plt.figure(figsize=(10, 8))
sns.barplot(x='importance', y='feature', data=feature_importances.head(10))
plt.title('ëœë¤í¬ë ˆìŠ¤íŠ¸ ëª¨ë¸ì˜ íŠ¹ì„± ì¤‘ìš”ë„ Top 10', fontsize=15, fontweight='bold')
plt.xlabel('íŠ¹ì„± ì¤‘ìš”ë„')
plt.ylabel('íŠ¹ì„±ëª…')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()

print("\nâœ… 3ë‹¨ê³„ ë¶„ì„ ì™„ë£Œ. ìƒê´€ê´€ê³„ê°€ ë†’ì€ ë³€ìˆ˜ë“¤ì„ ì œê±°í•˜ê³ ë„ ìœ ì˜ë¯¸í•œ ì˜ˆì¸¡ ëª¨ë¸ì„ ë§Œë“¤ì—ˆìŠµë‹ˆë‹¤.")
print("   ë‹¤ìŒ ë‹¨ê³„ì—ì„œëŠ” ìƒê´€ë¶„ì„ê³¼ ëœë¤í¬ë ˆìŠ¤íŠ¸ ê²°ê³¼ë¥¼ ì¢…í•©í•˜ì—¬ ìµœì¢… 'í†µí•© ì ìˆ˜'ë¥¼ ë§Œë“¤ê² ìŠµë‹ˆë‹¤.")
```

```python
# =======================================================
# 4ë‹¨ê³„: í†µí•© ê°€ì¤‘ì¹˜ ë° ìµœì¢… ì ìˆ˜ ëª¨ë¸ë§
# =======================================================

print("ğŸ”„ 4ë‹¨ê³„ ë¶„ì„ ì‹œì‘: í†µí•© ê°€ì¤‘ì¹˜ ë° ìµœì¢… ì ìˆ˜ ëª¨ë¸ë§")

# --- 1. ë‘ ë¶„ì„ ê²°ê³¼ì˜ ê°€ì¤‘ì¹˜ í†µí•© ---
print("\nâš–ï¸ ìƒê´€ë¶„ì„ê³¼ ëœë¤í¬ë ˆìŠ¤íŠ¸ ê²°ê³¼ì˜ ê°€ì¤‘ì¹˜ë¥¼ í†µí•©í•©ë‹ˆë‹¤...")

# 2ë‹¨ê³„ ìƒê´€ë¶„ì„ ê²°ê³¼ (ìƒìœ„ 15ê°œ)
corr_weights = corr_df_sorted.head(15)[['abs_correlation']].copy()
corr_weights.rename(columns={'abs_correlation': 'corr_score'}, inplace=True)

# 3ë‹¨ê³„ ëœë¤í¬ë ˆìŠ¤íŠ¸ ê²°ê³¼ (ìƒìœ„ 15ê°œ)
rf_weights = feature_importances.head(15).set_index('feature')[['importance']]
rf_weights.rename(columns={'importance': 'rf_score'}, inplace=True)

# ë‘ ê²°ê³¼ë¥¼ ë³‘í•©
integrated_weights_df = pd.merge(corr_weights, rf_weights, left_index=True, right_index=True, how='outer').fillna(0)

# ê° ì ìˆ˜ë¥¼ 0-1 ì‚¬ì´ë¡œ ì •ê·œí™”
integrated_weights_df['corr_norm'] = integrated_weights_df['corr_score'] / integrated_weights_df['corr_score'].sum()
integrated_weights_df['rf_norm'] = integrated_weights_df['rf_score'] / integrated_weights_df['rf_score'].sum()

# 50:50ìœ¼ë¡œ ìµœì¢… ê°€ì¤‘ì¹˜ ê³„ì‚°
integrated_weights_df['final_weight'] = (integrated_weights_df['corr_norm'] * 0.5) + (integrated_weights_df['rf_norm'] * 0.5)

# ìµœì¢… ê°€ì¤‘ì¹˜ë¥¼ ë‹¤ì‹œ 100%ë¡œ ì •ê·œí™”
integrated_weights_df['final_weight'] /= integrated_weights_df['final_weight'].sum()

# ìµœì¢… ê°€ì¤‘ì¹˜ ìˆœìœ¼ë¡œ ì •ë ¬
final_weights = integrated_weights_df[['final_weight']].sort_values('final_weight', ascending=False)

print("\n--- ğŸ† ìµœì¢… í†µí•© ê°€ì¤‘ì¹˜ Top 10 íŠ¹ì„± ---")
display(final_weights.head(10))

# --- 2. ìµœì¢… 'íˆ¬ì ì ì¬ë ¥ ì ìˆ˜' ê³„ì‚° ---
print("\nğŸ’¯ ë°±ë¶„ìœ„ìˆ˜(Percentile)ë¥¼ ì´ìš©í•´ ê° ì§€ì—­ì˜ ìµœì¢… ì ìˆ˜ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤...")

# ìµœì¢… ê°€ì¤‘ì¹˜ì— í¬í•¨ëœ íŠ¹ì„±ë“¤ë§Œ ì‚¬ìš©
scoring_features = final_weights.index.tolist()
df_merged['final_score'] = 0

for feature in scoring_features:
    # ê° ì§€ì—­ì´ ì „ì²´ì—ì„œ ìƒìœ„ ëª‡ %ì— ìœ„ì¹˜í•˜ëŠ”ì§€ë¥¼ ì ìˆ˜í™” (0~100ì )
    percentile_rank = df_merged[feature].rank(pct=True) * 100

    # í•´ë‹¹ íŠ¹ì„±ì˜ ê°€ì¤‘ì¹˜ë¥¼ ê³±í•˜ì—¬ ì ìˆ˜ ëˆ„ì 
    weight = final_weights.loc[feature, 'final_weight']
    df_merged['final_score'] += percentile_rank * weight

print("âœ… ìµœì¢… 'íˆ¬ì ì ì¬ë ¥ ì ìˆ˜' ê³„ì‚° ì™„ë£Œ!")

# --- 3. ìµœì¢… ì ìˆ˜ ë¶„í¬ í™•ì¸ ---
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(df_merged['final_score'], kde=True, bins=30, color='skyblue')
plt.title('ìµœì¢… íˆ¬ì ì ì¬ë ¥ ì ìˆ˜ ë¶„í¬', fontsize=15, fontweight='bold')
plt.xlabel('í†µí•© ì ìˆ˜ (0-100)')
plt.ylabel('ì§€ì—­ ìˆ˜')

plt.subplot(1, 2, 2)
sns.scatterplot(x='ë‹¹ì›”_ë§¤ì¶œ_ê¸ˆì•¡', y='final_score', data=df_merged, alpha=0.5)
plt.title('ì›” ë§¤ì¶œê³¼ ìµœì¢… ì ìˆ˜ ê´€ê³„', fontsize=15, fontweight='bold')
plt.xlabel('ë‹¹ì›” ë§¤ì¶œ ê¸ˆì•¡ (ë¡œê·¸ ìŠ¤ì¼€ì¼)')
plt.ylabel('í†µí•© ì ìˆ˜ (0-100)')
plt.xscale('log') # ë§¤ì¶œì•¡ì€ í¸ì°¨ê°€ í¬ë¯€ë¡œ ë¡œê·¸ ìŠ¤ì¼€ì¼ë¡œ í‘œí˜„
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()

print("\n--- ğŸ“ˆ ìµœì¢… ì ìˆ˜ ìƒìœ„ 10ê°œ ì§€ì—­ ---")
display(df_merged.sort_values('final_score', ascending=False)[['í–‰ì •ë™_ì½”ë“œ_ëª…', 'final_score', 'ìƒê¶ŒíšŒì „ìœ¨', 'ë‹¹ì›”_ë§¤ì¶œ_ê¸ˆì•¡']].head(10))

print("\nâœ… 4ë‹¨ê³„ ë¶„ì„ ì™„ë£Œ. ëª¨ë“  ì§€ì—­ì— ëŒ€í•´ 0ì ë¶€í„° 100ì ê¹Œì§€ì˜ ê°ê´€ì ì¸ 'íˆ¬ì ì ì¬ë ¥ ì ìˆ˜'ë¥¼ ë¶€ì—¬í–ˆìŠµë‹ˆë‹¤.")
print("   ì´ì œ ì´ ì ìˆ˜ë¥¼ í™œìš©í•˜ì—¬ ë§ˆì§€ë§‰ ë‹¨ê³„ì¸ 'í´ëŸ¬ìŠ¤í„°ë§ ë° ì „ëµ ìˆ˜ë¦½'ì„ ì§„í–‰í•©ë‹ˆë‹¤.")
```

```python
# =======================================================
# 5-1ë‹¨ê³„: ìµœì  í´ëŸ¬ìŠ¤í„° ê°œìˆ˜(k) ì‹¬ì¸µ íƒìƒ‰ (ì£¼ì„ ì¶”ê°€ ë²„ì „)
# =======================================================
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples
import matplotlib.cm as cm

print("ğŸ”„ 5-1ë‹¨ê³„ ë¶„ì„ ì‹œì‘: ìµœì  í´ëŸ¬ìŠ¤í„° ê°œìˆ˜(k) ì‹¬ì¸µ íƒìƒ‰")

# --- 1. í´ëŸ¬ìŠ¤í„°ë§ì„ ìœ„í•œ ë°ì´í„° ì¤€ë¹„ ---
clustering_features = [
    'final_score', 'ë‹¹ì›”_ë§¤ì¶œ_ê¸ˆì•¡', 'ë‹¹ì›”_ë§¤ì¶œ_ê±´ìˆ˜', 'ì—…ì¢…ë‹¤ì–‘ì„±ì§€ìˆ˜', 'ê±°ë˜ë³€ë™ì„±'
]
clustering_data = df_merged[clustering_features].fillna(0)
scaler = StandardScaler()
data_scaled = scaler.fit_transform(clustering_data)
print("âœ… í´ëŸ¬ìŠ¤í„°ë§ì„ ìœ„í•œ ë°ì´í„° ìŠ¤ì¼€ì¼ë§ ì™„ë£Œ")

# --- 2. ì—˜ë³´ìš°, ì‹¤ë£¨ì—£ ì ìˆ˜, ì‹¤ë£¨ì—£ í”Œë¡¯ ë¶„ì„ ---
k_range = range(2, 7)
wcss_scores = []
silhouette_scores = []

print("\nğŸ“Š í´ëŸ¬ìŠ¤í„° ê°œìˆ˜(k)ì— ë”°ë¥¸ ì„±ëŠ¥ ì§€í‘œ ê³„ì‚° ì¤‘...")

# ì‹¤ë£¨ì—£ í”Œë¡¯ì„ ê·¸ë¦¬ê¸° ìœ„í•œ Figure ì„¤ì •
fig, axes = plt.subplots(len(k_range), 2, figsize=(15, len(k_range) * 5))
fig.tight_layout(pad=4.0)

for i, k in enumerate(k_range):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(data_scaled)

    wcss_scores.append(kmeans.inertia_)

    silhouette_avg = silhouette_score(data_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print(f"   - k={k}: WCSS = {kmeans.inertia_:.2f}, í‰ê·  ì‹¤ë£¨ì—£ ì ìˆ˜ = {silhouette_avg:.4f}")

    # --- ì‹¤ë£¨ì—£ í”Œë¡¯ ì‹œê°í™” ---
    ax = axes[i][0]
    y_lower = 10
    sample_silhouette_values = silhouette_samples(data_scaled, cluster_labels)

    for j in range(k):
        jth_cluster_silhouette_values = sample_silhouette_values[cluster_labels == j]
        jth_cluster_silhouette_values.sort()

        size_cluster_j = jth_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_j

        color = cm.nipy_spectral(float(j) / k)
        ax.fill_betweenx(np.arange(y_lower, y_upper), 0, jth_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        ax.text(-0.05, y_lower + 0.5 * size_cluster_j, str(j))
        y_lower = y_upper + 10

    ax.set_title(f"k = {k}ì¼ ë•Œì˜ ì‹¤ë£¨ì—£ í”Œë¡¯", fontweight='bold')
    ax.set_xlabel("ì‹¤ë£¨ì—£ ê³„ìˆ˜ ê°’")
    ax.set_ylabel("í´ëŸ¬ìŠ¤í„° ë ˆì´ë¸”")
    ax.axvline(x=silhouette_avg, color="red", linestyle="--")
    ax.set_yticks([])

# --- ì—˜ë³´ìš° ë° ì‹¤ë£¨ì—£ ì ìˆ˜ êº¾ì€ì„  ê·¸ë˜í”„ ---
# ì—˜ë³´ìš° ì°¨íŠ¸
axes[0][1].plot(k_range, wcss_scores, 'o-', color='royalblue')
axes[0][1].set_title('ì—˜ë³´ìš° ë°©ë²• (Elbow Method)', fontsize=12)
axes[0][1].set_xlabel('í´ëŸ¬ìŠ¤í„° ê°œìˆ˜ (k)')
axes[0][1].set_ylabel('WCSS')
axes[0][1].grid(True, linestyle='--', alpha=0.6)
# ğŸš¨ ì£¼ì„ ì¶”ê°€ (k=4 ì§€ì )
axes[0][1].annotate('Elbow Point (k=4)', xy=(4, wcss_scores[2]), xytext=(4.5, wcss_scores[2] + 100),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8),
             fontsize=12, fontweight='bold')

# ì‹¤ë£¨ì—£ ì ìˆ˜ ì°¨íŠ¸
axes[1][1].plot(k_range, silhouette_scores, 'o-', color='green')
axes[1][1].set_title('í‰ê·  ì‹¤ë£¨ì—£ ì ìˆ˜', fontsize=12)
axes[1][1].set_xlabel('í´ëŸ¬ìŠ¤í„° ê°œìˆ˜ (k)')
axes[1][1].set_ylabel('ì ìˆ˜')
axes[1][1].grid(True, linestyle='--', alpha=0.6)
# ğŸš¨ ì£¼ì„ ì¶”ê°€ (ìµœê³ ì )
best_k_silhouette = k_range[np.argmax(silhouette_scores)]
axes[1][1].annotate(f'Best Score (k={best_k_silhouette})', xy=(best_k_silhouette, max(silhouette_scores)),
             xytext=(best_k_silhouette + 0.5, max(silhouette_scores) - 0.01),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8),
             fontsize=12, fontweight='bold')

# ë‚˜ë¨¸ì§€ ë¹ˆ subplot ìˆ¨ê¸°ê¸°
for i in range(2, len(k_range)):
    fig.delaxes(axes[i][1])

plt.suptitle('ìµœì  í´ëŸ¬ìŠ¤í„° ê°œìˆ˜(k) íƒìƒ‰ì„ ìœ„í•œ ì¢…í•© ë¶„ì„', fontsize=18, fontweight='bold', y=1.02)
plt.show()

# --- ìµœì¢… ë¶„ì„ ê²°ê³¼ ---
print(f"\n--- ë¶„ì„ ê²°ê³¼ ---")
print(f"ì—˜ë³´ìš° ì°¨íŠ¸ì—ì„œëŠ” k=3 ë˜ëŠ” k=4 ì§€ì ì—ì„œ ê¸°ìš¸ê¸°ê°€ ì™„ë§Œí•´ì§€ëŠ” 'íŒ”ê¿ˆì¹˜' í˜•íƒœê°€ ê´€ì°°ë©ë‹ˆë‹¤.")
print(f"ì‹¤ë£¨ì—£ ì ìˆ˜ëŠ” k={best_k_silhouette}ì¼ ë•Œ {max(silhouette_scores):.4f}ë¡œ ê°€ì¥ ë†’ì€ ê°’ì„ ê¸°ë¡í–ˆìŠµë‹ˆë‹¤.")
print(f"ì‹¤ë£¨ì—£ í”Œë¡¯ì„ ë³´ë©´, k=3ì¼ ë•Œ ê° í´ëŸ¬ìŠ¤í„°ì˜ ëª¨ì–‘ê³¼ ë‘ê»˜ê°€ ë¹„êµì  ê· ì¼í•˜ê³  í‰ê·  ì ìˆ˜(ë¶‰ì€ì„ ) ì´ìƒì˜ ë°ì´í„°ê°€ ë§ì•„ ì•ˆì •ì ì…ë‹ˆë‹¤.")
print(f"âœ… ê¸°ìˆ ì  ì§€í‘œë“¤ì„ ì¢…í•©í–ˆì„ ë•Œ, ìµœì ì˜ í´ëŸ¬ìŠ¤í„° ê°œìˆ˜ëŠ” **{best_k_silhouette}ê°œ**ë¡œ íŒë‹¨í•˜ëŠ” ê²ƒì´ ê°€ì¥ í•©ë¦¬ì ì…ë‹ˆë‹¤.")
print(f"   (ë‹¨, ë¹„ì¦ˆë‹ˆìŠ¤ ëª©ì ì— ë”°ë¼ 4ê°œ íŒ€ êµ¬ì„±ì„ ìœ„í•´ k=4ë¥¼ ì„ íƒí•  ìˆ˜ë„ ìˆìŠµë‹ˆë‹¤.)")
print("\në‹¤ìŒ ë‹¨ê³„ì—ì„œëŠ” ì´ ê²°ê³¼ë¥¼ ë°”íƒ•ìœ¼ë¡œ k=3ê³¼ k=4 ì‹œë‚˜ë¦¬ì˜¤ë¥¼ ë¹„êµ ë¶„ì„í•©ë‹ˆë‹¤.")
```

```python
# =======================================================
# 5-2ë‹¨ê³„: ìµœì¢… ì‹œë‚˜ë¦¬ì˜¤ ë¹„êµ ë° ì „ëµ ìˆ˜ë¦½ (ì¤‘ë³µ ì œê±° ìµœì¢…ì•ˆ)
# =======================================================
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Pandas ì¶œë ¥ ì˜µì…˜ ì„¤ì • (ìˆ«ì í˜•ì‹)
pd.options.display.float_format = '{:,.2f}'.format

print("ğŸ”„ 5-2ë‹¨ê³„ ë¶„ì„ ì‹œì‘: k=3 vs k=4 ì‹œë‚˜ë¦¬ì˜¤ ë¹„êµ ë¶„ì„")

# --- 1. í´ëŸ¬ìŠ¤í„°ë§ì„ ìœ„í•œ ë°ì´í„° ì¤€ë¹„ ---
clustering_features = [
    'final_score', 'ë‹¹ì›”_ë§¤ì¶œ_ê¸ˆì•¡', 'ë‹¹ì›”_ë§¤ì¶œ_ê±´ìˆ˜', 'ì—…ì¢…ë‹¤ì–‘ì„±ì§€ìˆ˜', 'ê±°ë˜ë³€ë™ì„±'
]
clustering_data = df_merged[clustering_features].fillna(0)
scaler = StandardScaler()
data_scaled = scaler.fit_transform(clustering_data)
pca = PCA(n_components=2)
pca_result = pca.fit_transform(data_scaled)

# --- 2. ë‘ ê°€ì§€ ì‹œë‚˜ë¦¬ì˜¤(k=3, k=4)ë¡œ í´ëŸ¬ìŠ¤í„°ë§ ìˆ˜í–‰ ë° ìƒì„¸ ë¶„ì„ ---
scenarios = {'k=3 (ê¸°ìˆ ì  ìµœì )': 3, 'k=4 (ë¹„ì¦ˆë‹ˆìŠ¤ ëª©ì  ë¶€í•©)': 4}
results = {}

for name, k in scenarios.items():
    print("\n" + "="*80)
    print(f"--- ì‹œë‚˜ë¦¬ì˜¤ ë¶„ì„: {name} ---")
    print("="*80)

    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(data_scaled)

    scenario_df = df_merged.copy()
    scenario_df['cluster'] = labels
    scenario_df['pca_x'] = pca_result[:, 0]
    scenario_df['pca_y'] = pca_result[:, 1]

    # --- ğŸ“Š í´ëŸ¬ìŠ¤í„°ë³„ í‰ê·  íŠ¹ì„± ë¶„ì„ ê²°ê³¼ ---
    print("\n--- ğŸ“Š í´ëŸ¬ìŠ¤í„°ë³„ í‰ê·  íŠ¹ì„± ë¶„ì„ ê²°ê³¼ ---")
    summary = scenario_df.groupby('cluster')[clustering_features].mean().sort_values('final_score', ascending=False).reset_index()

    tier_map_k = {}
    tiers = ['ğŸ’ í”„ë¦¬ë¯¸ì—„ ìƒê¶Œ', 'ğŸš€ ê³ ì„±ì¥ ì ì¬ ìƒê¶Œ', 'ğŸ“ˆ ì•ˆì • ì„±ì¥ ìƒê¶Œ', 'ğŸŒ± ì´ˆê¸° ì ì¬ ìƒê¶Œ'][0:k]
    for i in range(k):
        tier_map_k[summary.loc[i, 'cluster']] = tiers[i]

    summary['tier'] = summary['cluster'].map(tier_map_k)
    scenario_df['tier'] = scenario_df['cluster'].map(tier_map_k)
    display(summary)

    # --- ğŸ¯ ìµœì¢… ì „ëµ ë§¤íŠ¸ë¦­ìŠ¤ ë° ì• ìì¼íŒ€ ë¯¸ì…˜ ---
    print("\n--- ğŸ¯ ìµœì¢… ì „ëµ ë§¤íŠ¸ë¦­ìŠ¤ ë° ì• ìì¼íŒ€ ë¯¸ì…˜ ---")
    region_counts = scenario_df['tier'].value_counts().reset_index()
    region_counts.columns = ['tier', 'region_count']
    final_summary = pd.merge(summary, region_counts, on='tier')

    strategy_map_k = {
        'ğŸ’ í”„ë¦¬ë¯¸ì—„ ìƒê¶Œ': 'ìˆ˜ìµ ê·¹ëŒ€í™” ë° VIP ê´€ë¦¬',
        'ğŸš€ ê³ ì„±ì¥ ì ì¬ ìƒê¶Œ': 'ê³µê²©ì  ë§ˆì¼€íŒ… ë° ì ìœ ìœ¨ í™•ëŒ€',
        'ğŸ“ˆ ì•ˆì • ì„±ì¥ ìƒê¶Œ': 'ì„œë¹„ìŠ¤ ìµœì í™” ë° ê³ ê° ìœ ì§€',
        'ğŸŒ± ì´ˆê¸° ì ì¬ ìƒê¶Œ': 'ì €ë¹„ìš© íƒìƒ‰ ë° ê¸°íšŒ ë°œêµ´'
    }
    final_summary['strategy'] = final_summary['tier'].map(strategy_map_k)

    final_summary['í‰ê·  ì›”ë§¤ì¶œ(ì–µì›)'] = final_summary['ë‹¹ì›”_ë§¤ì¶œ_ê¸ˆì•¡'] / 1e8

    display_cols = ['tier', 'region_count', 'strategy', 'final_score', 'í‰ê·  ì›”ë§¤ì¶œ(ì–µì›)', 'ê±°ë˜ë³€ë™ì„±', 'ì—…ì¢…ë‹¤ì–‘ì„±ì§€ìˆ˜']
    display(final_summary[display_cols].rename(columns={'final_score': 'í‰ê·  ì ì¬ë ¥ ì ìˆ˜'}))

    # --- ğŸ† í´ëŸ¬ìŠ¤í„°ë³„ Top 5 íˆ¬ì ìš°ì„ ìˆœìœ„ ì§€ì—­ ---
    print("\n--- ğŸ† í´ëŸ¬ìŠ¤í„°ë³„ Top 5 íˆ¬ì ìš°ì„ ìˆœìœ„ ì§€ì—­ ---")
    for tier_name in final_summary['tier']:
        print(f"\n--- {tier_name} ---")

        # ğŸš¨ í•´ê²°ì±…: ì ìˆ˜ ìˆœìœ¼ë¡œ ì •ë ¬ í›„, í–‰ì •ë™ ê¸°ì¤€ìœ¼ë¡œ ì¤‘ë³µì„ ì œê±°í•˜ê³  ìƒìœ„ 5ê°œë¥¼ ì„ íƒ
        tier_df = scenario_df[scenario_df['tier'] == tier_name].sort_values('final_score', ascending=False)
        top_5 = tier_df.drop_duplicates(subset=['í–‰ì •ë™_ì½”ë“œ_ëª…']).head(5)

        top_5['ë‹¹ì›”_ë§¤ì¶œ_ê¸ˆì•¡(ì–µì›)'] = top_5['ë‹¹ì›”_ë§¤ì¶œ_ê¸ˆì•¡'] / 1e8
        display(top_5[['í–‰ì •ë™_ì½”ë“œ_ëª…', 'final_score', 'ë‹¹ì›”_ë§¤ì¶œ_ê¸ˆì•¡(ì–µì›)']].reset_index(drop=True))

    results[name] = {'df': scenario_df, 'summary': summary}


# --- 3. ì‹œë‚˜ë¦¬ì˜¤ ë¹„êµ ì‹œê°í™” ---
print("\n" + "="*80)
print("--- ğŸ¨ PCA ê¸°ë°˜ í´ëŸ¬ìŠ¤í„° ì‹œê°í™” ë¹„êµ ---")
print("="*80)

fig, axes = plt.subplots(1, 2, figsize=(18, 8))
fig.suptitle('k=3 vs k=4 í´ëŸ¬ìŠ¤í„°ë§ ì‹œë‚˜ë¦¬ì˜¤ ë¹„êµ', fontsize=20, fontweight='bold')

# k=3 ì‹œë‚˜ë¦¬ì˜¤ ì‹œê°í™”
ax = axes[0]
sns.scatterplot(x='pca_x', y='pca_y', hue='tier', data=results['k=3 (ê¸°ìˆ ì  ìµœì )']['df'],
                palette='viridis', alpha=0.8, s=50, ax=ax)
ax.set_title('ì‹œë‚˜ë¦¬ì˜¤ 1: k=3 (ê¸°ìˆ ì  ìµœì )', fontsize=15)
ax.set_xlabel(f'ì£¼ì„±ë¶„ 1 (ì„¤ëª…ë ¥: {pca.explained_variance_ratio_[0]:.1%})')
ax.set_ylabel(f'ì£¼ì„±ë¶„ 2 (ì„¤ëª…ë ¥: {pca.explained_variance_ratio_[1]:.1%})')
ax.legend(title='í´ëŸ¬ìŠ¤í„° ë“±ê¸‰')
ax.grid(True, linestyle='--', alpha=0.6)

# k=4 ì‹œë‚˜ë¦¬ì˜¤ ì‹œê°í™”
ax = axes[1]
sns.scatterplot(x='pca_x', y='pca_y', hue='tier', data=results['k=4 (ë¹„ì¦ˆë‹ˆìŠ¤ ëª©ì  ë¶€í•©)']['df'],
                palette='plasma', alpha=0.8, s=50, ax=ax)
ax.set_title('ì‹œë‚˜ë¦¬ì˜¤ 2: k=4 (ë¹„ì¦ˆë‹ˆìŠ¤ ëª©ì  ë¶€í•©)', fontsize=15)
ax.set_xlabel(f'ì£¼ì„±ë¶„ 1 (ì„¤ëª…ë ¥: {pca.explained_variance_ratio_[0]:.1%})')
ax.set_ylabel(f'ì£¼ì„±ë¶„ 2 (ì„¤ëª…ë ¥: {pca.explained_variance_ratio_[1]:.1%})')
ax.legend(title='í´ëŸ¬ìŠ¤í„° ë“±ê¸‰')
ax.grid(True, linestyle='--', alpha=0.6)

plt.show()


# --- 5. ìµœì¢… ê²°ë¡  ë° ê¶Œì¥ì‚¬í•­ ---
print("\n" + "="*80)
print("--- ğŸ’¡ ìµœì¢… ê²°ë¡  ë° ê¶Œì¥ì‚¬í•­ ---")
print("="*80)
print("k=3 ì‹œë‚˜ë¦¬ì˜¤ëŠ” ê¸°ìˆ ì ìœ¼ë¡œëŠ” ê°€ì¥ ì•ˆì •ì ì´ì§€ë§Œ, 'í”„ë¦¬ë¯¸ì—„' ê·¸ë£¹ê³¼ 'ê³ ì„±ì¥' ê·¸ë£¹ì´ í•˜ë‚˜ë¡œ ë¬¶ì—¬ ìˆì–´ ì„¸ë¶„í™”ëœ ì „ëµ ìˆ˜ë¦½ì— í•œê³„ê°€ ìˆìŠµë‹ˆë‹¤.")
print("k=4 ì‹œë‚˜ë¦¬ì˜¤ëŠ” ì‹¤ë£¨ì—£ ì ìˆ˜ëŠ” ë‹¤ì†Œ ë‚®ì§€ë§Œ, ë¹„ì¦ˆë‹ˆìŠ¤ ëª©í‘œì¸ 4ê°œ íŒ€ êµ¬ì„±ì— ë¶€í•©í•˜ë©°, ê° ê·¸ë£¹ì˜ íŠ¹ì„±ì´ ë” ëª…í™•í•˜ê²Œ êµ¬ë¶„ë˜ì–´ ë§ì¶¤í˜• ì „ëµì„ ìˆ˜ë¦½í•˜ê¸°ì— ë” ìš©ì´í•©ë‹ˆë‹¤.")
print("\nâœ… **ìµœì¢… ê¶Œì¥**: ë¹„ì¦ˆë‹ˆìŠ¤ í™œìš©ì„±ê³¼ ì „ëµì˜ êµ¬ì²´ì„±ì„ ê³ ë ¤í•˜ì—¬ **k=4 ì‹œë‚˜ë¦¬ì˜¤ë¥¼ ì±„íƒ**í•˜ëŠ” ê²ƒì„ ê°•ë ¥íˆ ê¶Œì¥í•©ë‹ˆë‹¤.")
print("\nğŸ‰ ëª¨ë“  ë¶„ì„ì´ ì„±ê³µì ìœ¼ë¡œ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤! ì´ ê²°ê³¼ë“¤ì„ ë°”íƒ•ìœ¼ë¡œ ë°œí‘œ ìë£Œë¥¼ êµ¬ì„±í•˜ì‹œë©´ ë©ë‹ˆë‹¤.")
```

```python
# =======================================================
# 5-3ë‹¨ê³„ (ì‹¬í™”): ì£¼ì„±ë¶„(PCA) ì˜ë¯¸ í•´ì„
# =======================================================

print("ğŸ” PCAì˜ ê° ì¶•(ì£¼ì„±ë¶„)ì´ ì–´ë–¤ ì˜ë¯¸ë¥¼ ê°–ëŠ”ì§€ ë¶„ì„í•©ë‹ˆë‹¤.")

# PCA ëª¨ë¸ì€ ì´ì „ 5-2ë‹¨ê³„ì—ì„œ í•™ìŠµëœ 'pca' ë³€ìˆ˜ë¥¼ ê·¸ëŒ€ë¡œ ì‚¬ìš©í•©ë‹ˆë‹¤.
# pca.components_ ëŠ” ê° ì£¼ì„±ë¶„(PC1, PC2)ì´ ì›ë˜ íŠ¹ì„±ë“¤ê³¼ ì–¼ë§ˆë‚˜ ê´€ë ¨ìˆëŠ”ì§€ë¥¼ ë³´ì—¬ì¤ë‹ˆë‹¤.
components_df = pd.DataFrame(pca.components_,
                             columns=clustering_features,
                             index=['ì£¼ì„±ë¶„ 1 (ê°€ë¡œì¶•)', 'ì£¼ì„±ë¶„ 2 (ì„¸ë¡œì¶•)'])

print("\n--- ì£¼ì„±ë¶„ë³„ íŠ¹ì„± ê¸°ì—¬ë„ ---")
# ê° ì¶•ì—ì„œ ì ˆëŒ“ê°’ì´ ê°€ì¥ í° íŠ¹ì„±ë“¤ì´ í•´ë‹¹ ì¶•ì˜ ì˜ë¯¸ë¥¼ ê²°ì •í•©ë‹ˆë‹¤.
display(components_df)

# ê° ì£¼ì„±ë¶„ì˜ ì˜ë¯¸ í•´ì„
pc1_description = components_df.T.sort_values('ì£¼ì„±ë¶„ 1 (ê°€ë¡œì¶•)', ascending=False)
pc2_description = components_df.T.sort_values('ì£¼ì„±ë¶„ 2 (ì„¸ë¡œì¶•)', ascending=False)

print("\n--- í•´ì„ ê²°ê³¼ ---")
print(f"ê°€ë¡œì¶•(ì£¼ì„±ë¶„ 1)ì— ê°€ì¥ í° ì˜í–¥ì„ ì£¼ëŠ” íŠ¹ì„± Top 3:")
print(pc1_description.head(3))
print(f"\nì„¸ë¡œì¶•(ì£¼ì„±ë¶„ 2)ì— ê°€ì¥ í° ì˜í–¥ì„ ì£¼ëŠ” íŠ¹ì„± Top 3:")
print(pc2_description.head(3))

print("\nğŸ’¡ ìœ„ ê²°ê³¼ë¥¼ ë°”íƒ•ìœ¼ë¡œ ê° ì¶•ì˜ ë¹„ì¦ˆë‹ˆìŠ¤ ì˜ë¯¸ë¥¼ ì •ì˜í•˜ê³  ë°œí‘œì— í™œìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.")
```
